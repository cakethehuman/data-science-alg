```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn.metrics as metrics
import sklearn.model_selection as ms
import sklearn.ensemble as ens
import sklearn.preprocessing as pre
```

```python
rf = ens.RandomForestClassifier(n_estimators=100)

df = pd.read_csv('adult_data.csv')
```

```python
gender = pre.LabelEncoder()
gender.fit(df["sex"])
df["gender_code"] = gender.transform(df["sex"])
```

```python
df
```

```python
education = pre.LabelEncoder()
education.fit(df["education"])
df["education_code"] = education.transform(df["education"])

race = pre.LabelEncoder()
race.fit(df["race"])
df["race_code"] = race.transform(df["race"])

workclass = pre.LabelEncoder()
workclass.fit(df["workclass"])
df["workclass_code"] = workclass.transform(df["workclass"])

occupation = pre.LabelEncoder()
occupation.fit(df["occupation"])
df["occupation_code"] = occupation.transform(df["occupation"])

relationship = pre.LabelEncoder()
relationship.fit(df["relationship"])
df["relationship_code"] = relationship.transform(df["relationship"])

native = pre.LabelEncoder()
native.fit(df["native_country"])
df["native_country_code"] = native.transform(df["native_country"])
```

```python
df
```

```python
df["label"] = df["income"]




x = df[["age","education_num","capital_gain", "capital_loss", "hours-per-week","education_code","race_code","workclass_code","occupation_code","relationship_code","native_country_code"]]
y = df["label"]
```

```python
df
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(x, y, test_size=0.2)
```

```python
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("score akurasi = ",metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
print(rf.feature_importances_)
```
