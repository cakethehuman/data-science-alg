```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn.metrics as metrics
import sklearn.model_selection as ms
import sklearn.ensemble as ens
import sklearn.preprocessing as pre
```

```python
df = pd.read_csv('loan_approval_dataset.csv')
```

```python
df
```

```python
df.drop('loan_id', axis=1, inplace=True)
```

```python
df
```

```python
df
```

```python
df["education"].unique()
```

```python
df["education"] = df["education"].map({"NotGraduate": 0, "Graduate": 1})
df["self_employed"] = df["self_employed"].map({"No": 0, "Yes": 1})
df["loan_status"] = df["loan_status"].map({"Rejected": 0, "Approved": 1})
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
df_correlation = df.corr()
df_correlation["loan_status"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(df_correlation, annot=True)

plt.show()
```

```python
rf = ens.RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
x = df[['cibil_score', 'loan_amount', 'commercial_assets_value', 'education', 'self_employed']]
y = df[['loan_status']]   
```

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
r2 = rf.score(x_test, y_test)
r2
```
