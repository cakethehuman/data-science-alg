```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.model_selection as ms
```

```python
df = pd.read_csv("bensin.csv")
df.corr()
```

```python
liter = df[['Liter']]
kiiometer = df[['Kilometer']]

X_train,x_test,y_train,y_test = ms.train_test_split(liter,kiiometer,test_size=0.2,random_state=0)

```

```python
plt.scatter(X_train,y_train,edgecolors='r')
plt.xlabel("Liter")
plt.ylabel("kilometer")
plt.title("fuel burn")
x1 = np.linspace(0,45)
y1 = 4 + 7 * x1
plt.plot(x1,y1)
plt.show()
```

```python
x1 = np.linspace(0,45)
y1 = 4 + 7 * x1

plt.plot(x1,y1)
plt.show()
```

```python
import sklearn.linear_model as ml

model = ml.LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x1 = np.linspace(0,45)
y1 =  10.641 + 6.45 * x1

plt.scatter(X_train,y_train,edgecolors='r')
plt.xlabel("Liter")
plt.ylabel("kilometer")
plt.title("fuel burn")

plt.plot(x1,y1)
plt.show()
```

```python
r2 = model.score(x_test,y_test)
r2
```

```python
score = model.predict([[60]])
print(score)
```
