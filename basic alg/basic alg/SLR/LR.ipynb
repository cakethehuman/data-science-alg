```python
import pandas as pd
import numpy as np

import sklearn.model_selection as ms
import matplotlib.pyplot as plt

import sklearn.linear_model as ml

model = ml.LinearRegression()
df = pd.read_csv("bensin2.csv")
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(['RPM'], axis = 1) 
```

```python
df
```

```python
df.corr()
```

```python
x = df[['Liter','Penumpang','Kecepatan','Suhu']]
y = df[['Kilometer']]

X_train,x_test,y_train,y_test = ms.train_test_split(x,y,test_size = 0.1, random_state = 0)
model.fit(X_train,y_train)
```

```python
print(f"this is {model.coef_}")
model.intercept_
```

```python
data = np.array([[20,1,42,29]])
output = model.predict(data)
output
```

```python
r2 = model.score(x_test,y_test)
r2
```
