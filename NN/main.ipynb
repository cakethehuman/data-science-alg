```python
import pandas as pd
import sklearn.neural_network as nn 
import sklearn.model_selection as ms
import sklearn.preprocessing as pp
```

```python
df = pd.read_csv("datatraining.csv")
```

```python
df
```

```python
x = df[['Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio']]
y = df["Occupancy"]
```

```python
x.describe()
```

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x, y, test_size=0.2)
```

```python
x_train.count()
```

```python
scl = pp.StandardScaler(copy=True, with_mean=True, with_std=True)
scl.fit(x_train)
x_train = scl.transform(x_train)
x_test = scl.transform(x_test)
```

```python
mlp = nn.MLPClassifier(hidden_layer_sizes=(3),max_iter=5)
mlp.fit(x_train, y_train)
```

```python
y_pred = mlp.predict(x_test)
print(y_pred)
```

```python
import sklearn.metrics as met
print(met.classification_report(y_test, y_pred))
```
